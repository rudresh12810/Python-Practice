### Task 1
Here we have two lists (year and GDP).
Please create a line plot as proper labels and title to the figure.

Year = [1950, 1960, 1970, 1980, 1990]

GDP = [50.8, 65.9, 75.8, 92.6, 125.8]



```python
Year = [1950, 1960, 1970, 1980, 1990]
GDP = [50.8, 65.9, 75.8, 92.6, 125.8]

plt.plot(Year,GDP)
plt.xlabel("Year")
plt.ylabel("GDP")
plt.title("Change in GDP")
plt.show()
```

## Task 5
Please read mtcars dataset and create a histogram of mpg (mile per gallon) feature with proper lables and title. Give it an appropriate color.


```python
mtcars = pd.read_excel("mtcars.xlsx")
mtcars.head()
```

```python
plt.hist(mtcars.mpg,color='maroon')
plt.xlabel('Miles per gallon')
plt.ylabel('Frequency')
plt.title('Distribution of MPG in mtcars')
plt.show()
```

## Task 7
Please use mtcars dataset and create a scatter plot between mpg (mile per gallon) and wt (weight) feature with proper lables and title. Change the marker style to hexagon, add face color as blue, edge color as Navy and size 40.




```python
plt.scatter(mtcars['mpg'],mtcars['wt'],marker = 'H', 
            edgecolor = 'Navy', facecolor = 'blue',s=40)
plt.xlabel('Miles per gallon')
plt.ylabel('Weight')
plt.title('Relationship between Miles per gallon and Weight')
plt.show()
```

## Task 8
Please use mtcars dataset and create a scatter plot between mpg (mile per gallon) and wt (weight). Scatter plot should be between am 0 and 1.



```python
# Create separate dataframes for am = 0 (automatic) and am = 1 (manual)
am0 = mtcars[mtcars['am'] == 0]
am1 = mtcars[mtcars['am'] == 1]

plt.scatter(am0['wt'], am0['mpg'], color='red', label='Automatic', marker = 'H', edgecolor= 'black')
plt.scatter(am1['wt'], am1['mpg'], color='blue', label='Manual', marker = 'D', edgecolor= 'black')


plt.xlabel('Weight')
plt.ylabel('Miles per gallon')
plt.title('Relationship between weight and mpg by transmission type')

plt.legend(loc = "lower left")

plt.show()
```

## Task 9
Create a density plot from mtcars dataset using weight column. Add appropriate label and title.


```python
sns.kdeplot(mtcars['wt'],fill = True)
plt.title("Density Plot of Weight in mtcars Dataset")
plt.xlabel("Weight")
plt.ylabel("Density")
plt.show()
```
