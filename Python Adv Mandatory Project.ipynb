```python
def Calculate_median(dataset,column) : # here I'm defining a function for Calculate median of any Column which i want 
    if column not in dataset.columns:
        print("Column not Found in DataFrame")
        
    median_value = dataset[column].median()
    return median_value

result = Calculate_median(df,'total_bedrooms') # here i'm calling the function and passing the dataframe and column as argument 
print(result)


#Cross Check
print("median of total_beddrooms :",df['total_bedrooms'].median())
```

#### 10. 
Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
df.head()
```

```python
def bedroom_size_categorizer(column): # here I'm defining a function for Categorize Column data 
    if column <=10:
        return 'Small'
    elif column >=11 and column <1000:
        return 'Medium'
    else:
        return 'Large'
        

# by using apply()  method, i categorize the 'total_bedrooms' column and assign to the new column 
df['total_bedroom_size'] = df['total_bedrooms'].apply(bedroom_size_categorizer)

# here i'm changing the dtype as Category 
df['total_bedroom_size'] = df['total_bedroom_size'].astype(('category'))

print(df['total_bedroom_size'].value_counts())
```
